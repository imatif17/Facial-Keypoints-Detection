## Dataset Class

```python
class FaceKeyPointDataset(Dataset):
    
    def __init__(self,csv_file,root_dir,transforms = False):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.keypoint_frame = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transforms = transforms
    
    def __len__(self):
        return len(self.keypoint_frame)
    
    def __getitem__(self,idx):
        img = mpimg.imread(os.path.join(self.root_dir,self.keypoint_frame.iloc[idx,0]))
        
        if(img.shape[2] == 4):
            img = img[:,:,0:3]
            
        key_pts = self.keypoint_frame.iloc[idx,1:].as_matrix()
        key_pts = key_pts.astype('float').reshape(-1,2)
        sample = {'image': img,'keypoints': key_pts}
        if (self.transforms):
            sample = self.transforms(sample)
        return sample
```

```python
face_dataset = FaceKeyPointDataset('data/training_frames_keypoints.csv','data/training/')
print('Length of dataset : '+ str(len(face_dataset)))
```

## Transforms

```python
class Normalize(object):
    def __call__(self,sample):
        img, key_pts = sample['image'],sample['keypoints']
        gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
        
        gray = gray / 255.0
        key_pts = (key_pts - 100) / 50.0
        return {'image': gray,'keypoints':key_pts}
    
        
class Rescale(object):
    def __init__(self,output_size):
        assert isinstance(output_size,(int,tuple))
        self.output_size = output_size
        
    def __call__(self,sample):
        img = sample['image']
        key_pts = sample['keypoints']
        h, w = img.shape[:2]
        if isinstance(self.output_size, int):
            if h > w:
                new_h, new_w = self.output_size * h / w , self.output_size
            else:
                new_h, new_w = self.output_size, self.output_size * w / h
                
        else:
            new_h, new_w = self.output_size
        
        new_h, new_w = int(new_h),int(new_w)
        
        img = cv2.resize(img,(new_w,new_h))
        
        key_pts = key_pts * [new_w/w, new_h/h]
        return {'image':img,'keypoints':key_pts}
    
class RandomCrop(object):
    def __init__(self,output_size):
        
        assert isinstance(output_size, (int, tuple))
        if isinstance(output_size, int):
            self.output_size = (output_size, output_size)
        else:
            assert len(output_size) == 2
            self.output_size = output_size
            
    def __call__(self,sample):
        img, key_pts = sample['image'],sample['keypoints']
        
        h,w = img.shape[:2]
        new_h,new_w = self.output_size
        
        top = np.random.randint(0,h - new_h)
        left = np.random.randint(0,w - new_w)
        
        img = img[top : top + new_h, left : left + new_w]
        key_pts = key_pts - [left, top]
        return {'image':img,'keypoints':key_pts}
    
class ToTensor(object):
    def __call__(self,sample):
        img, key_pts = sample['image'],sample['keypoints']
        if(len(img.shape) == 2):
            img = img.reshape(img.shape[0],img.shape[1],1)
        
        img = img.transpose((2,0,1))
        return {'image':torch.from_numpy(img),'keypoints':torch.from_numpy(key_pts)}
        
```

## Testing Transforms

```python
rescale = Rescale(100)
crop = RandomCrop(50)
composed = transforms.Compose([Rescale(250),
                               RandomCrop(224)])

idx = 19
sample = face_dataset[idx]
for i, tx in enumerate([rescale,crop,composed]):
    transformed_sample = tx(sample)
    ax = plt.subplot(1, 3, i + 1)
    plt.tight_layout()
    ax.set_title(type(tx).__name__)
    show_image(transformed_sample['image'], transformed_sample['keypoints'])

plt.show()
```

## Creating Transform Dataset

```python
data_transform = transforms.Compose([Rescale(250),RandomCrop(224),Normalize(),ToTensor()])
transformed_dataset = FaceKeyPointDataset('data/training_frames_keypoints.csv','data/training/',transforms = data_transform)
```
